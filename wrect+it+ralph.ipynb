```python
fh = open(r'C:\Users\dell\Desktop\材料\wreck it ralph.txt')

text = fh.read()

fh.close()

print(len(text))
print(text)
```

```python
text = text.replace(',', '')
text = text.replace("'", '')
text = text.replace('.', '')
text = text.replace('!', '')
text = text.replace('?', '')
text = text.replace('-', '')
text = text.replace('\n',' ')


start = 0
end = 0

idioms = []
while end < len(text):
    if text[end] == ' ':
        idioms.append(text[start:end])
        start = end + 1
    end += 1

while '' in idioms:
    idioms.remove('')
print(idioms)
```
